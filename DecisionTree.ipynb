```python
import pandas as pd #For data analysis
import numpy as np #Numerical calculations
from sklearn.datasets import load_iris #scikit library-->polular in ML--> split a dataset into training and testing sets
from sklearn import tree #To build a tree
```

```python
iris=load_iris() #Loading iris database
```

```python
print(iris.feature_names)
```

```python
iris
```

```python
#Convert nD array to dataframe
irisdata=iris.data #defining the data part of table
colnam=iris.feature_names #Defining the columns part of table
df=pd.DataFrame(data=irisdata,columns=colnam)
```

```python
df
```

```python
df.to_csv("iriss.csv",index=False)
```

```python
#Read the csv file
dff=pd.read_csv("/content/iriss.csv")
type(dff)
```

```python
# Assign data and column
X=dff #X-->feature names(['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
Y=iris.target #Y-->The target i.e(0,1,2)
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=42)
# test_size--> split a dataset into training and testing sets
# random_state-->set a seed for the random number generator(code again with the same seed, you'll get the same train-test split)
```

```python
# Train classifier
clf=tree.DecisionTreeClassifier(criterion="entropy",max_depth=3,min_samples_split=5)# Defining decision tree classifier
#criterion-->quality of a split(less entropy,more Info. gain)
#
clf=clf.fit(X_train,Y_train)# Train data on new data and new target
prediction=clf.predict(X_test)# Assign removed data as input
```

```python
from sklearn.metrics import accuracy_score
```

```python
#print("Original Labels",X_test)
#print("Labels Predicted",prediction)

print("Train data accuarcy: ",accuracy_score(y_true=Y_train,y_pred=clf.predict(X_train)))
print("Test data accuracy: ",accuracy_score(y_true=Y_test,y_pred=prediction))
```
